```python
my_schedule.loc[:,'شماره درس']= my_schedule[['شماره درس']].astype('int64')
my_schedule.loc[:,'گروه درس']= my_schedule[['گروه درس']].astype('int64')

my_schedule.loc[:,'شماره درس']= my_schedule[['شماره درس']].astype(str)
my_schedule.loc[:,'گروه درس']= my_schedule[['گروه درس']].astype(str)
```

```python
def zero_lesson_group (column) :
    if len(column) < 2:
        return '0' + column 
    return column
my_schedule.loc[:,'گروه درس'] = my_schedule['گروه درس'].apply(zero_lesson_group)
```

```python
my_schedule['lesson_id'] = my_schedule['شماره درس'] + my_schedule['گروه درس']
```
